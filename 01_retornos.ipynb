```python
import yfinance as yf
msft = yf.Ticker("MSFT")
hist = msft.history(period="max")
```

```python
import numpy as np 
import pandas as pd
```

```python
prices =  hist['Close']
```

```python
prices.loc['2020':].plot(figsize = (16, 6))
```

### Simple Return 
$$r_t = \frac{P_t - P_{t-1}}{P_{t-1}}$$

```python
returns = prices.pct_change().dropna()
```

```python
returns.plot(kind='hist', bins = 100, figsize = (16, 6))
```

### Logarithmic Return 
$$r_t = ln(P_t) - ln(P_{t-1})$$
$$r_t = ln(P_t/P_{t-1})$$


```python
logreturns = np.log(prices.shift(1)/prices)
```

```python
logreturns.plot(kind='hist', bins =100, figsize=(16, 6))
```
